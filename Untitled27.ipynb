```python
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt
import csv
import matplotlib.pyplot as plt
from mne.decoding import CSP
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data =  pd.read_csv(r"pa_3_labeled.csv")
print(data.head())
```

```python

fs = 250
final_data=np.empty([1,22,201])

final_label = []

for i in range(288):
    final_label.append(data["class"][i*201])


    lowcut = 8  # Hz
    highcut = 30  # Hz

    order = 4

    nyq = 0.5 * fs

    low = lowcut / nyq
    high = highcut / nyq

    b, a = butter(order, [low, high], btype='band')


    data_to_filter = data.iloc[i:i+201, 5:27]

    i = i+201


    temp = data_to_filter


    filtered_data = pd.DataFrame(
        filtfilt(b, a, temp, axis=0),
        columns=temp.columns,
        index=temp.index
    )
    

    filtered_data = filtered_data - filtered_data.mean(axis=1).values.reshape(-1, 1)

    normalized_data = (filtered_data - filtered_data.mean()) / filtered_data.std()

    temp = data_to_filter.T
    
    final_data = np.concatenate((final_data, np.expand_dims(temp, axis=0)), axis=0)



final_data = final_data[1:]
```

```python
import tkinter as tk
from tkinter import ttk
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from mne.decoding import CSP
from sklearn.model_selection import train_test_split

# Assuming X_train, X_test, y_train, y_test are already defined from your data preparation

def map_eeg_to_activity(eeg_class):
    # Map EEG class to activity
    if eeg_class == "Tongue":
        return "Sleep"
    elif eeg_class == "Left Hand":
        return "Eat"
    elif eeg_class == "Right Hand":
        return "Drink Water"
    else:
        return "Go to Bathroom"

def classify_activity():
    # Preprocessing pipeline
    csp = CSP(n_components=4, reg=None, log=True, norm_trace=False)
    csp.fit(X_train, y_train)
    X_train_csp = csp.transform(X_train)
    X_test_csp = csp.transform(X_test)

    pca = PCA(n_components=2)
    X_train_pca = pca.fit_transform(X_train_csp)
    X_test_pca = pca.transform(X_test_csp)

    scaler = MinMaxScaler()
    X_train_norm = scaler.fit_transform(X_train_pca)
    X_test_norm = scaler.transform(X_test_pca)

    # Classifier
    clf = RandomForestClassifier(n_estimators=100)
    clf.fit(X_train_norm, y_train)
    y_pred = clf.predict(X_test_norm)

    # Map predicted EEG class to activity
    predicted_activity = map_eeg_to_activity(eeg_combo.get())

    # Compute accuracy
    accuracy = accuracy_score(y_test, y_pred)
    result_label.config(text=f"Predicted Activity: {predicted_activity}\nAccuracy: {accuracy:.2f}")

# Create main window
root = tk.Tk()
root.title("Motor Imagery Activity Prediction")

# Create labels and buttons
instruction_label = ttk.Label(root, text="Select an EEG class to imagine:")
instruction_label.pack(pady=10)

eeg_combo = ttk.Combobox(root, values=["Tongue", "Left Hand", "Right Hand"])
eeg_combo.pack()

classify_button = ttk.Button(root, text="Classify", command=classify_activity)
classify_button.pack(pady=10)

result_label = ttk.Label(root, text="Predicted Activity: ")
result_label.pack()

# Run the UI loop
root.mainloop()
```
